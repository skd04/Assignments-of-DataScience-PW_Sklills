### Q1. What is Flask Framework? What are the advantages of Flask Framework?

### Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask, redirect

# call the app
app = Flask(__name__)

# Create the route
@app.route("/simple")
def helloWorld(): 
    return "<h1> Hello World </h1>"

@app.route("/") # Redirect 
def redi_h():
    return redirect("/simple")

# run the flask
if __name__ == '__main__':
    app.run(host="0.0.0.0", debug=True)
```

### Q3. What is App routing in Flask? Why do we use app routes?

_App routing in Flask means linking URLs to specific functions using @app.route(). It's like giving directions when a user visits a URL, Flask knows which function to run. We use app routes to organize web pages, handle user requests, and create APIs.Just like a map, Flask knows where to go when someone clicks on a link or visits a page, making website building simpler._

### Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details: 

#### Company Name: ABC Corporation
#### Location: India
#### Contact Detail: 999-999-9999

```python
from flask import Flask, redirect

# Call the app
app = Flask(__name__)

# Create the Welcome route
@app.route("/welcome")
def helloWorld(): 
    return "<h1> Welcome to ABC Corporation </h1>"

@app.route("/") # For showing contact details
def redi_h():
    return """
    <h3> Company Name: ABC Corporation </h3>
    <h3> Location: India </h3>
    <h3> Contact Detail: 999-999-9999 </h3>
    """

# Run the flask
if __name__ == '__main__':
    app.run(host="0.0.0.0", debug=True)
```

### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

 _The url_for() function is used in Flask for URL Building._

```python
from flask import Flask, redirect, url_for

# Call the app
app = Flask(__name__)

# Create the Welcome route
@app.route("/home")
def home(): 
    return """<h1> Welcome to Testing page <br> Here I'm trying to learn Data Science because I love this field but... </h1> 
    <br> For more click on <a href='{}'>HERE</a>.""".format(url_for('contacts'))

# Connect the page 
@app.route("/contacts")
def contacts():
    return """
    <h4>Hi I'm a B.Com Honours Student I need a Job, but my family need my support so I'm doing a Job which is not give me anything for my career.<br>
    There did not have any PaySlip or Experience letter. Before that I'm work in a Consultancy firm because of they don't give any piece of paper.<br>
    I cannot mention that I've experience in accounting work and GST works. Yes now i forgot those works because of not in touch of those.<br> If possible 
    or any kind of job opportunity can you please contact me via my mail (sdjobymail@gmail.com). Thanks and regards</h4>
    <a href='{}'>Back</a>.""".format(url_for('home'))

@app.route("/") # For showing contact details
def redi_h():
    return redirect(url_for('contacts'))

# Run the flask
if __name__ == '__main__':
    app.run(host="0.0.0.0", debug=True)
```
