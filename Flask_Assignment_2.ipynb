### Q1. Explain GET and POST methods.

_GET and POST are methods used in web communication. GET is like asking for information. GET is like asking for a webpage by clicking a link. It's visible in the URL. POST is like submitting a form with your information. It's not visible in the URL and is safer for sensitive details like passwords._

### Q2. Why is request used in Flask?

_In Flask, "request" is used to handle incoming HTTP requests from clients. It allows Flask applications to access data submitted by clients. It helps your Flask app understand what the user wants to do, like submitting a form or getting a webpage.It's basically the way your Flask app listens to what the user is asking for and responds accordingly._

### Q3. Why is redirect() used in Flask?

_"redirect()" in Flask is used to send users to a different URL within the application. It's handy for navigation and managing user flow.It's used after they do something, like filling out a form or logging in. Instead of staying on the same page, redirect() takes them to the right place._

### Q4. What are templates in Flask? Why is the render_template() function used?

_Templates in Flask are files that contain static and dynamic HTML content  like ready-made forms. You can add your own text and pictures to them._

  _The render_template() function in Flask is used to render HTML templates and pass dynamic data to them. It takes a template's name and any needed data, then combines them to create a complete web page.It's an easy way to make your web pages look good and have the right information on them._

```python
from flask import Flask, redirect, jsonify, request

app = Flask(__name__)

@app.route("/api", methods = ['POST'])
def api():
    if request.method == 'POST':
        msg = request.json["message"]
        return jsonify({"message": msg})

@app.route("/")
def redi_api():
    return redirect("/api")

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
